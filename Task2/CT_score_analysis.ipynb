```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dat = pd.read_csv('CT_score_results.csv')
```

```python
dat
```

```python
dat['Matching probability score'] = dat['ATAC2GEX']
dat['Cell type match score'] = dat['CT_match']
```

```python
plot_dat = pd.melt(dat, id_vars=['Team Name'], value_vars=['Matching probability score', 'Cell type match score'])
```

```python
plot_dat
```

```python
###### sns.set_style("white")
#sns.set(rc={'figure.figsize':(25,8)}, style='white')
sns.set(font_scale = 1.6, style='white')
sns.catplot(x="Team Name", y="value", hue='variable',  data=plot_dat, palette='Set2', height=5, aspect=8/5, legend=False, s=10)
sns.despine()
plt.xticks(rotation=90)
plt.legend(loc=(0.7,0.8), frameon=False)
plt.savefig('ATAC2GEX_vs_CT_match.pdf', bbox_inches="tight")
```
